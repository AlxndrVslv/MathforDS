### Задание 1

В магазин привезли устройства с 3-х разных предприятий. 

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта. 

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
### YOUR CODE/SOLUTION HERE
# Обозначим события:
# An - устройство было взято из n-го предприятия;
# Bn - устройство из n-го предприятия оказалось высшего сорта;

PA1 = 0.2
PA2 = 0.3
PA3 = 0.5

# Вероятность того, что устройство оказалось в/с при условии, что оно взято из первого предприятия:
PB1_if_A1 = 0.1
# Вероятность того, что устройство оказалось в/с при условии, что оно взято из второго предприятия:
PB2_if_A2 = 0.05
# Вероятность того, что устройство оказалось в/с при условии, что оно взято из третьего предприятия:
PB3_if_A3 = 0.2

# Вероятность того, что устройство в/с взято из первого предприятия:
PA1B1 = PB1_if_A1 * PA1
PA2B2 = PB2_if_A2 * PA2
PA3B3 = PB3_if_A3 * PA3

# Обшая вероятность выбрать устройство высшего сорта:
P = PA1B1 + PA2B2 + PA3B3
print(f'Вероятность того, что продукция окажется высшего сорта P = {P:.2%}')
```

### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
### YOUR SOLUTION HERE
import numpy as np
```

```python
# Счетчик событий A1, A2, A3
nA1 = nA2 = nA3 = 0

# Счетчик попарных срабатываний событий и всех событий в совокупности
nA1A2 = nA1A3 = nA2A3 = nA1A2A3 = 0

# Счетчик повторений эксперимента
n_tries = 10**5

for i in range(n_tries):
    A1 = A2 = A3 = False

    # Сгенерируем случайные события подбрасывания монет
    coin_1 = np.random.choice([0, 1])
    coin_2 = np.random.choice([0, 1])
    coin_3 = np.random.choice([0, 1])
    
    # Посчитаем кол-во успешных событий A1, A2, A3
    if coin_1 == coin_2: 
        nA1 += 1
        A1 = True
    if coin_2 == coin_3: 
        nA2 += 1
        A2 = True
    if coin_1 == coin_3: 
        nA3 += 1
        A3 = True
    
    # Посчитаем кол-во успешных попарных событий A1A2, A1A3, A2A3
    if A1 & A2: nA1A2 += 1
    if A1 & A3: nA1A3 += 1
    if A2 & A3: nA2A3 += 1

    # Посчитаем кол-во, когда все события сработали в совокупности
    if A1 & A2 & A3: nA1A2A3 += 1

# События попарно независимы, когда P(AB) = P(A) * P(B)

# Вероятности событий A1, A2, A3:
PA1 = nA1 / n_tries    
PA2 = nA2 / n_tries
PA3 = nA3 / n_tries

# Вероятности попарных событий A1A2, A1A3, A2A3:
PA1A2 = nA1A2 / n_tries
PA1A3 = nA1A3 / n_tries
PA2A3 = nA2A3 / n_tries

# Вероятность совокупного события A1A2A3:

PA1A2A3 = nA1A2A3 / n_tries

print(f'Вероятность срабатывания событий A1 и A2 = {PA1A2:.2%}; Произведение вероятностей событий A1 и A2 = {(PA1 * PA2):.2%}')
print(f'Вероятность срабатывания событий A1 и A3 = {PA1A3:.2%}; Произведение вероятностей событий A1 и A3 = {(PA1 * PA3):.2%}')
print(f'Вероятность срабатывания событий A2 и A3 = {PA2A3:.2%}; Произведение вероятностей событий A2 и A3 = {(PA2 * PA3):.2%}\n')
print(f'Вероятность совокупного срабатывания всех событий = {PA1A2A3:.2%}; Произведение вероятностей событий A1, A2, A3 = {(PA1 * PA2 * PA3):.2%}')
```

<tr>Из эксперимента видно, что:
    <li>для попарных событий вероятность одновременного срабатывания двух событий приблизительно равна произведению вероятностей каждого из событий (при увеличении числа итераций эксперимента разница должна нивелироваться);</li>
    <li>для совокупного события данное условие не выполняется;</li>
</tr>
Соответственно, в совокупности эти 3 события не являются независимымы. Но являются таковыми попарно.

### Задание 3


Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения. 

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```python
### YOUR CODE HERE
from scipy import stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(nrows = 5, ncols = 3, figsize = (12, 10), sharey = True)
fig.tight_layout()

def norm_distr(ax, loc, scale):
    nrm = st.norm(loc = loc, scale = scale)
    nrm_rvs = sorted(nrm.rvs(size = 100))
    ax.plot(nrm_rvs, nrm.pdf(x = nrm_rvs), label = f'loc = {loc}\nscale = {scale}')
    ax.legend()

def expon_distr(ax, loc, scale):
    exp = st.expon(loc = loc, scale = scale)
    exp_rvs = sorted(exp.rvs(size = 100))
    ax.plot(exp_rvs, exp.pdf(x = exp_rvs), label = f'loc = {loc}\nscale = {scale}')
    ax.legend()

def t_distr(ax, loc, scale):
    df = 100000
    t = st.t(df = df, loc = loc, scale = scale)
    t_rvs = sorted(t.rvs(size = 100))
    ax.plot(t_rvs, t.pdf(x = t_rvs), label = f'loc = {loc}\nscale = {scale}')
    ax.legend()

ax[0, 0].set_title('Нормальное распределение')
ax[0, 1].set_title('Экспоненциальное распределение')
ax[0, 2].set_title('Распределение Стьюдента')

norm_distr(ax[0, 0], 0, 1)
norm_distr(ax[1, 0], 0, 5)
norm_distr(ax[2, 0], 0, 0.5)
norm_distr(ax[3, 0], 5, 5)
norm_distr(ax[4, 0], 5, 0.5)

expon_distr(ax[0, 1], 0, 1)
expon_distr(ax[1, 1], 0, 5)
expon_distr(ax[2, 1], 0, 0.5)
expon_distr(ax[3, 1], 5, 5)
expon_distr(ax[4, 1], 5, 0.5)

t_distr(ax[0, 2], 0, 1)
t_distr(ax[1, 2], 0, 5)
t_distr(ax[2, 2], 0, 0.5)
t_distr(ax[3, 2], 5, 5)
t_distr(ax[4, 2], 5, 0.5)

plt.show()
```

<tr>Из графиков видно, что:
    <li>параметр loc отвечает за расположение центрирования;</li>
    <li>параметр scale отвечает за коэффициент растяжения/сжатия графиков около центра;</li>
</tr>
